```python
import pandas as pd 

bios = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv")
display(bios.info())
# Filtering data
display(bios.loc[bios["height_cm"] > 200]) 
display(bios.loc[bios["height_cm"] > 200, ["name", "height_cm", "NOC"]])
display(bios[(bios["height_cm"] > 150) & (bios["born_city"] == "Delhi")])  # Filtering on multiple conditions
```

```python
import pandas as pd 

bios = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv")
# dataset["column_name"].str.contains("substring", case) method checks each entry in "column_name" to see if it has the "substring" provided. 
# The case argument makes this check case sensitivity in boolean value.
display(bios[bios["name"].str.contains("hans|david", case = False)])  # Example of regex filtering
display(bios[bios["name"].str.contains("hans|david", case = False, regex = False)])  # regex = False disables regex which returns no value
display(display(bios[bios["born_country"].isin(["FRA", "NOR", "AUS", "USA", "RUS"]) & (bios["name"].str.startswith("V"))]))
# Another method of filtering data is using query() functions
display(bios.query("born_country == 'GBR' and born_city == 'Birmingham'"))  
```
