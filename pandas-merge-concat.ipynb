Merging and Concatonating in Pandas

```python
import pandas as pd

nocs = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/noc_regions.csv")
bios = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv")
bios_new = bios.copy()
nocs.head()
# merge() method updates the content of two DataFrame by merging them together. It is used by specifying the columns beside the index 
# If a column is present in both left and right DataFrame, the merge() method will automatically add a suffix(A_x, A_y etc.) to the column name.
# we can specify the suffixes to be used using the suffixes parameter. Can be used on duplicate columns.
bios_new = pd.merge(bios, nocs, left_on="born_country", right_on="NOC", how="left", suffixes=('_bios', '_nocs'))
display(bios_new.head(2))
bios_new.rename(columns={'region': 'born_country_full'}, inplace=True)
display(bios_new.head(2))
display(bios_new[bios_new["NOC_bios"] != bios_new["born_country_full"]][['name', 'NOC_bios', 'born_country_full']].head(5))
```

```python
import pandas as pd

nocs = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/noc_regions.csv")
bios = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv")
bios_new = bios.copy()
usa = bios[bios["born_country"] == "USA"].copy()
gbr = bios[bios["born_country"] == "GBR"].copy()
# concat() method is used to concatenate two or more DataFrames. It is used by specifying the DataFrames to be concatenated as a list.
new_df = pd.concat([usa, gbr])
display(new_df.head(2))
display(new_df.tail(2))
```

```python
import pandas as pd

nocs = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/noc_regions.csv")
bios = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv")
results = pd.read_parquet("results.parquet")
display(results)
# another way to merge two dataframes using merge() method is by specifying the common columns to be used for merging
combined_df = pd.merge(results, bios, on="athlete_id", how="left")
display(combined_df)
```
