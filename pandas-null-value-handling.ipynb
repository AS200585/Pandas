Handling Null Values in Pandas

```python
import pandas as pd
import numpy as np
coffee = pd.read_csv('coffee.csv')
print(coffee.head())
coffee.loc[[3, 5, 8, 11], "Units Sold"] = np.nan
# print(coffee)
print(coffee.isna().sum())  # check for missing values(NaN)
display(coffee.fillna(500))  # fill missing values(NaN) with the a provided value
display(coffee.fillna(coffee["Units Sold"].mean()).head())
# interpolate() function fills missing or NaN values in the backward direction using interpolation method
display(coffee["Units Sold"].interpolate())
# display(coffee)
coffee["Units Sold"] = coffee["Units Sold"].interpolate()
display(coffee)
# coffee.dropna()  # drop rows with missing values
display(coffee["Units Sold"].isna())  # check for missing values(NaN)
display(coffee["Units Sold"].notna())  # check for non-missing values
```
