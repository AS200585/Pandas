Adding and Removing Columns in Pandas

```python
import pandas as pd

coffee = pd.read_csv("coffee.csv")
# Adding new columns to dataset
coffee["price"] = 3.59
display(coffee.head())
```

```python
import numpy as np
import pandas as pd

coffee = pd.read_csv("coffee.csv")
coffee["price"] = 3.59
# where () function enablesus to search, filter, and apply conditions to elements of an array, returning specific elements based on condition provided.
coffee['new_price'] = np.where(coffee['Coffee Type'] == "Espresso", 2.99, 1.99)
display(coffee)
# drop() method removes the specified row or column. 
# By default, the drop() method removes the specified row.
# By specifying the column axis ( axis='columns' ), the drop() method removes the specified column. 
# By specifying the row axis ( axis='index' ), the drop() method removes the specified row.
display(coffee.drop(0).head())
display(coffee.drop(columns=["Day"]).tail())
coffee.drop(columns = ["Day"], inplace= True)  # In order to actually modify the dataset
display(coffee)
# copy() method is used to create a copy of the original dataset
# the method creates a new list containing the same elements as the original list but maintains its own identity in memory.
coffee_new = coffee.copy()
coffee_new["new_price2"] = 6
display(coffee.tail())
display(coffee_new)
```

```python
import numpy as np
import pandas as pd

coffee = pd.read_csv("coffee.csv")
coffee_new = coffee.copy()
coffee_new["new_price2"] = 4.79
coffee.drop(columns=["price", "new_price"], inplace= True, errors="ignore")  # errors="ignore" parameter will skip dropping columns if they don't exist.
coffee_new["revenue"] = coffee_new["Units Sold"] * coffee_new["new_price2"]
display(coffee_new.head())
coffee_new.rename(columns={"new_price2" : "price"}, inplace= True)
# another method is - coffee = coffee_new.rename(columns= {"new_price2" : "price"}) 
display(coffee_new.head())
```

```python
import pandas as pd

bios = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv")
display(bios.head(5))
bios_new = bios.copy()
# to_datetime() converts a scalar, array, Series or DataFrame to a pandas datetime object or format
# Setting errors="coerce" means that any invalid date values will be converted to NaT (Not a Time) instead of causing an error, helpful in data cleaning.
bios_new["born_datetime"] = pd.to_datetime(bios_new["born_date"], errors="coerce")  #  assigningresults to a new column, born_datetime
display(bios_new.info())
bios_new["born_year"] = bios_new["born_datetime"].dt.year
display(bios_new[["name", "born_year"]].head())
bios_new["born_year"] = bios_new["born_datetime"].dt.is_leap_year  # returns boolean value to check if it is leap year or not
display(bios_new[["name", "born_year"]].head(15))
```

```python
import pandas as pd

bios = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv")
# apply() method is one of the most common methods of data preprocessing
# It allows you to perform operations on each element, row, or column of a DataFrame, making it versatile for data transformation and manipulation.
bios["height_category"] = bios["height_cm"].apply(lambda x: "Short" if x < 165 else ("Average" if x < 180 else "Tall"))
bios["weight_category"] = bios["weight_kg"].apply(lambda x: "Light" if x < 65 else ("Moderate" if x < 80 else "Heavy" if x < 100 else "Overweight"))
display(bios.head())
display(bios.tail())

def athlete_category(row):
    if row['height_cm'] < 165 and row['weight_kg'] < 70:
        return "Light-Weight"
    elif row['height_cm'] < 180 and row['weight_kg'] < 85:
        return "Middle-Weight"
    else:
        return "Heavy-Weight"
    
bios["Category"] = bios.apply(athlete_category, axis = 1)  # axis = 1 means apply function to each row, axis = 0 means apply function to each column
display(bios.head())
```
