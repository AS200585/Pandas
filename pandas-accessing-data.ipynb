```python
import pandas as pd 

coffee = pd.read_csv("coffee.csv")
coffee.sample(8, random_state = 1)  # Display random values from the dataset
# random_state=1 is set, the same "random" rows will be selected each time you run the code.
# Reproducibility means getting the same results each time you run a piece of code, as long as the inputs and environment remain the same.
```

```python
import pandas as pd 

coffee = pd.read_csv("coffee.csv")
display(coffee.loc[[1, 3, 6, 9]])  # Display the row with a particular index 
display(coffee.loc[3:11])
display(coffee.loc[4:7, ["Day", "Coffee Type"]])
"""
If you use .iloc[1:3], it will return the rows at indices 1 and 2, but not 3.

If you use .loc[1:3], it will return rows labeled 1, 2, and 3.
"""
display(coffee.iloc[:12, 2:])
```

```python
coffee.index = coffee["Units Sold"]
coffee.head()
```

```python
import pandas as pd 

coffee = pd.read_csv("coffee.csv")
coffee.index = coffee["Day"]
display(coffee.loc["Monday" : "Thursday", "Coffee Type"])
```

```python
import pandas as pd 

coffee = pd.read_csv("coffee.csv")
# Modyfying values at a particular index
coffee.loc[3, "Units Sold"] = 4
coffee.loc[1, "Coffee Type"] = "Cappucino"
display(coffee)
coffee.loc[5:10, "Units Sold"] = [5, 6, 7, 8, 9, 10]  # Modifying multiple values at multiple indexes
coffee.loc[6:9, "Coffee Type"] = "Frappucino"
display(coffee)
```

```python
# Accessing single values from a dataset
import pandas as pd 
coffee = pd.read_csv("coffee.csv")
display(coffee.at[7, "Coffee Type"])  
display(coffee.iat[8, 2])  # Accessing values by index
```

```python
import pandas as pd 
coffee = pd.read_csv("coffee.csv")
display(coffee["Coffee Type"])
# sort_values(column_name, ascending = True or False) used to sort values in dataset 
display(coffee.sort_values("Units Sold", ascending = True))
# 0 for False and 1 for True
display(coffee.sort_values(["Units Sold", "Coffee Type"], ascending=[0, 1]))
```

```python
import pandas as pd 

# Performance issues occur during iteration
coffee = pd.read_csv("coffee.csv")
#  iterrows() method generates an iterator object of the DataFrame, allowing us to iterate each row in the DataFrame
for row, index in coffee.iterrows():
    print(index)
    print(row)
    print("\n")
```
