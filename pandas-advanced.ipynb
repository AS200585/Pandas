Pandas Advanced Functions

```python
import pandas as pd

coffee = pd.read_csv("coffee.csv")
bios = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv")
print(coffee.head())
print(coffee.describe())
# shift() is a pandas function that moves the column down by a specified number of rows
coffee["yesterday_sales"] = coffee["Units Sold"].shift(1) # shift moves the column down by 1
display(coffee)
coffee["pct_change"] = (coffee["Units Sold"] / coffee["yesterday_sales"]) * 100
display(coffee)
#  rank() is a pandas function that ranks the values in a column.
bios["height_rank"] = bios["height_cm"].rank(ascending=False)  
bios.sort_values(["height_rank"], inplace=True)
display(bios)
# cumsum() is a pandas function that calculates the cumulative sum of a column.
coffee["cumulative_sales"] = coffee["Units Sold"].cumsum()
display(coffee)
# rolling() is a pandas function that calculates the rolling average of a column. It takes in the window size as an argument.
# rolling average is the average of a data set updated continuously to include all the data in the set until that point.
latte = coffee[coffee["Coffee Type"] == "Latte"].copy()
latte["latte_3day_cumulative"] = latte["Units Sold"].rolling(3).sum()
espresso = coffee[coffee["Coffee Type"] == "Espresso"].copy()
espresso["espresso_3day_cumulative"] = espresso["Units Sold"].rolling(3).sum()
display(latte, espresso)
```
